## Phase 1: Data Automation (Lesson 03)
**🎯 The Goal:** Load the raw CSV and automatically filter out corrupted or impossible data.

**📄 Documentation Task:** Below this cell, add a Markdown cell explaining the specific "dirty data" issues you found in the raw dataset (e.g., negative marks) and how your Pandas code fixed them.

**📊 PPT Slide 2 (Data Strategy):** Show a screenshot of the dirty data vs. your clean data. Explain why "garbage data in" means "garbage AI predictions out."

```python
print("--- Loading and Cleaning Data ---")
# 1. Load the raw dataset
df_raw = pd.read_csv('master_markbook.csv')
print(f"Total rows before cleaning: {len(df_raw)}")

# 2. Drop any rows that have missing data (NaN)
df_clean = df_raw.dropna().copy()

# 3. Filter out impossible marks (must be between 0 and 100)
# TODO: Replace 'Maths_Advanced' and 'Physics' with your chosen subjects if different.
df_clean = df_clean[
    (df_clean['Maths_Advanced'].between(0, 100)) &
    (df_clean['Physics'].between(0, 100)) &
    (df_clean['SE_Final'].between(0, 100))
]

print(f"Total rows after cleaning: {len(df_clean)}")
```

---
## Phase 2: The Baseline AI (Lessons 01 & 02)
**🎯 The Goal:** Build a basic AI that predicts the final mark using only **one** subject.

**📄 Documentation Task:** Explain the 5 stages of the ML Pipeline (Define, Collect, Prepare, Train, Evaluate) in the context of this project.

**📊 PPT Slide 1 (The Problem):** Introduce Alex's missing exam scenario. Explain why splitting data into 80% Training and 20% Testing stops the AI from "cheating."

```python
print("--- Training Level 1 AI ---")
# 1. Extract ONE input subject (X1) and the Target (y)
X1 = df_clean[['Maths_Advanced']].values
y = df_clean['SE_Final'].values

# 2. Split the data 80% Training / 20% Testing
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)

# 3. Create and train the empty AI brain
model_1 = LinearRegression()
model_1.fit(X1_train, y_train)

# 4. Predict on the test data and calculate RMSE
predictions_1 = model_1.predict(X1_test)
rmse_1 = np.sqrt(mean_squared_error(y_test, predictions_1))

print(f"Level 1 AI RMSE: {rmse_1:.2f} marks error")
```

## Phase 3: Mathematical Proof (Lesson 04)
**🎯 The Goal:** Demystify the "black box" by extracting the exact formula the AI learned.

**📄 Documentation Task:** Write down your specific $y = wx + b$ formula. Explain what the weight (slope) means in plain English (e.g., "For every 1 extra mark in Maths, the Final Exam score goes up by X marks").

```python
print("--- Extracting the Formula ---")
# 1. Extract the learned weights and bias
weight = model_1.coef_[0]
bias = model_1.intercept_
print(f"The AI's Secret Formula: y = {weight:.2f}x + {bias:.2f}\n")

# 2. Plot the real student data as blue dots
plt.figure(figsize=(7, 4))
plt.scatter(X1_test, y_test, color='steelblue', alpha=0.6, label='Actual Students')

# 3. Plot the AI's prediction line in red
plt.plot(X1_test, predictions_1, color='red', linewidth=2, label='AI Line of Best Fit')

plt.title('Level 1 AI: Predicting Final Exam Marks')
plt.xlabel('Maths Advanced Score')
plt.ylabel('Software Engineering Final Score')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## Phase 4: The Object-Oriented Engine (Lesson 06)
**🎯 The Goal:** Package your algorithm inside a professional Python Class to meet HSC OOP requirements.

**📄 Documentation Task:** Draw a formal "Class Diagram" mapping out the attributes and methods of your `MarkPredictor`. Embed the image here.

**📊 PPT Slide 4 (Architecture):** Display your Class Diagram. Explain why Object-Oriented code is more secure and maintainable than loose scripts.

```python
# 1. Define the Blueprint (Class)
class MarkPredictor:
    def __init__(self):
        # Encapsulate the sklearn model safely inside the object
        self.model = LinearRegression()
        
    def fit(self, X, y):
        # Custom method to train the hidden model
        self.model.fit(X, y)
        return self
        
    def predict(self, X):
        # Custom method to return predictions
        return self.model.predict(X)

print("--- Training OOP Engine ---")
my_ai = MarkPredictor()
my_ai.fit(X1_train, y_train)
oop_rmse = np.sqrt(mean_squared_error(y_test, my_ai.predict(X1_test)))
print(f"OOP Engine RMSE: {oop_rmse:.2f} marks error")
```

## Phase 5: Scaling Up to Level 2 (Lesson 07)
**🎯 The Goal:** Make the AI smarter by using multiple subjects and Standardizing the data.

**📄 Documentation Task:** Explain why Feature Scaling is mathematically necessary when mixing subjects that have different standard deviations.

**📊 PPT Slide 5 (Level 2 AI):** Compare your Level 1 RMSE to your Level 2 RMSE to prove to the board that adding a second feature improved accuracy.

```python
print("--- Training Level 2 AI (Multi-Subject) ---")
# 1. Extract TWO input subjects
X2 = df_clean[['Maths_Advanced', 'Physics']].values

# 2. Split the new data
X2_train, X2_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# 3. Apply Feature Scaling (fit_transform on train, transform on test)
scaler = StandardScaler()
X2_train_scaled = scaler.fit_transform(X2_train)
X2_test_scaled = scaler.transform(X2_test)

# 4. Train the OOP Engine on the SCALED data
level_2_ai = MarkPredictor()
level_2_ai.fit(X2_train_scaled, y_train)

# 5. Evaluate the smarter AI
predictions_2 = level_2_ai.predict(X2_test_scaled)
rmse_2 = np.sqrt(mean_squared_error(y_test, predictions_2))
print(f"Level 2 AI RMSE: {rmse_2:.2f} marks error")
```

## Phase 6: The Ethics Audit (Lessons 08 & 12)
**🎯 The Goal:** Convert scores to a Pass/Fail classification and test for Disparate Impact (The 80% Rule).

**📄 Documentation Task:** Write your Ethics Report. Discuss the Robodebt scandal and explain why you must legally mandate a "Human-in-the-Loop" to verify these AI predictions.

**📊 PPT Slide 6 (Ethics & Conclusion):** Present your Bias Audit results. Give your final recommendation for what Alex's mark should be!

```python
print("--- Running Bias Audit (80% Rule) ---")

# 1. Define two demographics
group_a = df_clean[df_clean['Physics'] > 70]          # STEM Students
group_b = df_clean[df_clean['Modern_History'] > 70]   # Humanities Students

# 2. Calculate Pass Rates (Score >= 50)
group_a_pass_rate = (group_a['SE_Final'] >= 50).mean()
group_b_pass_rate = (group_b['SE_Final'] >= 50).mean()

# 3. Calculate Ratio
disparate_impact_ratio = group_b_pass_rate / group_a_pass_rate

print(f"Group A Pass Rate: {group_a_pass_rate:.1%}")
print(f"Group B Pass Rate: {group_b_pass_rate:.1%}")
print(f"Disparate Impact Ratio: {disparate_impact_ratio:.2f}")

# 4. Check Legal Threshold
if disparate_impact_ratio < 0.8:
    print("\nWARNING: Disparate Impact detected. The system favors Group A.")
else:
    print("\nAudit Passed: No significant bias detected between these groups.")
```

## Phase 7: Overfitting Check (Lesson 09)
**🎯 The Goal:** Use Cross-Validation to prove the AI isn't just memorizing the training data.

**📄 Documentation Task:** Explain what "Overfitting" means and why your Cross-Validation score proves your AI is ready to be used on next year's students.

```python
print("--- Cross-Validation Check ---")
# 1. Scale the ENTIRE dataset for testing
X2_all_scaled = scaler.fit_transform(X2)

# 2. Run 5-Fold validation
cv_scores = cross_val_score(
    LinearRegression(), 
    X2_all_scaled, 
    y, 
    cv=5, 
    scoring='neg_root_mean_squared_error'
)

# 3. Calculate mean (sklearn returns negative scores, so we invert them)
mean_cv_rmse = -cv_scores.mean()

print(f"Original Test RMSE: {rmse_2:.2f}")
print(f"5-Fold CV RMSE:     {mean_cv_rmse:.2f}")
print("If these numbers are close, your AI successfully generalized!")
```

## Phase 8: The Logic Gatekeeper (Lesson 10)
**🎯 The Goal:** Write a hard-coded safety check using Decision Tree logic to block bad data.

**📄 Documentation Task:** Draw a Decision Tree flowchart showing how this Yes/No gatekeeper works and embed it here.

```python
# 1. Define the Logic Gate function
def check_data_reliability(attendance_percentage):
    """Refuses to predict marks for students with severe chronic absenteeism."""
    
    if attendance_percentage < 50.0:
        print("ACCESS DENIED: Student attendance is too low. Data is unreliable.")
        return False
    else:
        print("ACCESS GRANTED: Data is reliable for AI prediction.")
        return True

# 2. Test the gatekeeper
print("Testing student with 45% attendance:")
check_data_reliability(45.0)

print("\nTesting student with 92% attendance:")
check_data_reliability(92.0)
```
